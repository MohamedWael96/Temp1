```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px

data = {
    'month' :['Jan','Feb','Mar','Apr','May','June','July'],
    'sales': [2000,2200,2300,2400,2500,2100,1890],
    'Profit':[400,600,300,500,800,408,950],
    'Region':['North','South','East','west','North','west','South']
}
df= pd.DataFrame(data)
df
```

```python
region_sales = df.groupby('Region')['sales'].sum()
region_sales
```

```python

plt.pie(region_sales, labels=region_sales.index, autopct='%1.1f%%')
plt.title('Sales by Region')
plt.show()
```

```python
plt.figure(figsize=(8,7))
sns.barplot(x='month',y='sales',data=df, palette='viridis')
plt.title('Sales by Month')
plt.xlabel('Month(M#)')
plt.ylabel('Sales($)')
plt.show()
```

```python
plt.bar(df['month'],df['sales'])
sns.barplot(x='month',y='sales',data=df, palette='Reds')
plt.show()
```

```python
plt.barh(df['month'],df['sales'])
plt.show()
```
